### part1

```python
with open("names.txt") as f:
    deta=f.readlines()
    
```

```python
#strip():delite /n. or use slicing: word[:-1]
#set:delete repeated argu

deta_clean =sorted(set([word.strip().lower()for word in deta]))
deta_clean[:9]
```

### part2 

```python
#defined anagram for a word

def anagram(word):
    return "".join(sorted(word))
```

```python
anagram_dic = {}
for word in deta_clean:
    ana=anagram(word)
    if ana in anagram_dic:
        anagram_dic[ana].append(word)
    else:
        anagram_dic[ana]=[word]
anagram_dic
```

### part3 

```python
a=max([len(key)for key in anagram_dic.keys()])+1
a
```

```python
cole=0                        #to check number of items

for counter in range(1,a):
    anagram_counter=0
    for k,v in anagram_dic.items():
        if len(k)==counter:
            anagram_counter +=len(v)
    print(counter,anagram_counter)
    cole +=anagram_counter    #to check number of items
    
len(deta_clean) == cole       #to check number of items
```
